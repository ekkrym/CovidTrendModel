### Relative improvements in MAE, median AE, RC  and WIS 

```python
RI_, WIS, coverage = evaluation(countries, date_limits=['2020-04-01','2021-12-15'])
```

```python
WIS["$RI_{WIS}$"] = (WIS["baseline_WIS"]-WIS["forecast_WIS"])/(WIS["baseline_WIS"]+1) 
coverage["RC"] = (coverage["forecast"]-coverage["baseline"])/(coverage["baseline"])
RI = RI_.merge(WIS, on=["country"], how="inner").merge(coverage, on=["country"], how="inner") 
```

```python
RI_.columns
```

```python
error_names = RI.columns[1:] 
RI = RI.sort_values(by=error_names[0])
fig,ax = plt.subplots(1,4,figsize=(12,14),sharey=True,dpi=900)
RI.index = RI["country"]
RI = RI.rename(columns={'$RI_{MAE}$':"RMAE",'$RI_{MedianAE}$':'RmedianAE', "$RI_{WIS}$":"RWIS"})
RI[["RMAE"]].plot.barh(ax=ax[0], grid=True, title= "",color="b",alpha=0.6) 
RI[['RmedianAE' ]].plot.barh(ax=ax[1],grid=True,  title= "",color="g",alpha=0.6)  
RI[["RC"]].plot.barh(ax=ax[2],grid=True, title= "", color="black",alpha=0.5)  
RI[["RWIS"]].plot.barh(ax=ax[3],grid=True, title= "", color="red",alpha=0.5, )   
ax[3].set_xlim([-0.5,0.5])
fig.tight_layout() 
plt.subplots_adjust(wspace=0, hspace=0) 

plt.savefig('rmae_rest.pdf')
```
